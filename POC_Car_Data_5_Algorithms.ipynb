# **Replacing Missing Values**

```python
df['beygir'] = df['beygir'].fillna(df.mean().iloc[0])
df
```

```python
df.isnull().sum()
```

# **Label Encoding**

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Marka'] = label_encoder.fit_transform(df['Marka'])
df['Yakit'] = label_encoder.fit_transform(df['Yakit'])
df['Sanziman'] = label_encoder.fit_transform(df['Sanziman'])
df['Model'] = label_encoder.fit_transform(df['Model'])
df
```

# **X,Y Split**

```python
x = df.iloc[:,0:12]
x
```

```python
y = df.iloc[:,12]
y
```

# **TRAIN TEST SPILIT**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.25, random_state=10)
```

```python
y_test
```

# **Algorithms**

## **MLR**

```python
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()
```

```python
std_reg.fit(x_train,y_train)
```

```python
y_predict = std_reg.predict(x_test)
y_predict
```

```python
score = std_reg.score(x_test, y_test)
score
```

### **Plotting**

```python
sns.regplot(x=y_test,y=y_predict,ci=None,color="Yellow")
plt.plot(x_train,std_reg.predict(x_train),color="green")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.show()
```

## **Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
X_poly = poly_reg.fit_transform(x)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

### **Plotting**

```python
plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x)),color='red')
plt.title('Car Data')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.show()
```

## **KNN**

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
knn_model.fit(x_train, y_train)
```

```python
y2_pred= knn_model.predict(x_test)
y2_pred
```

```python
score = knn_model.score(x_test, y_test)
score
```

## **SVM**

```python
from sklearn.svm import SVR
SVR_model = SVR(kernel='rbf')
SVR_model.fit(x_train, y_train)
```

```python
y3_predict = SVR_model.predict(x_test)
y3_predict
```

```python
score = SVR_model.score(x_test, y_test)
score
```

## **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 3, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_predict_4 = regressor.predict(x_test)
y_predict_4
```
